## 1.1 Les modèles génératifs

Le principal objectif des modèles génératifs est de créer à partir de données du monde réel, de nouvelles données qui ressemblent à celles du monde réel.
Les domaines d'applications sont tout aussi nombreux que variés, ils vont de la génération d'images de synthèse pour le cinéma, à la production de données massives pour de l'apprentissage d'algorithmes de machine learning en passant par l'élaboration de nouvelles formules de médicaments.

Dans le domaine de la production d'images de haute qualité, ceci dans de nombreux domaines, les modèles  Generative Adversarial Networks (GANs), Variational Autoencoders (VAEs), Normalizing Flows ont déjà démontré toute leur efficacité.

Les modèles de diffusion, plus récemment, viennent eux aussi s'ajouter à cette liste de modèles de génération d’images de haute qualité.

## 1.2 Architectures de modèles génératifs


Ces dernières années, trois architectures majeures ont joué un rôle central dans les avancées des modèles génératifs :
* les réseaux antagonistes génératifs (GANs)
* les flux normalisants (Normalizing Flows)
* les modèles de diffusion


<br/>Ces modèles sont populaires pour plusieurs raisons, les GANs ont longtemps été à la pointe de nombreuses applications, les flux normalisants se distinguent par leur capacité unique à modéliser avec précision la densité des données mais ces dernières années ce sont les modèles de diffusion qui ont connu un développement considérable.
  
<br/>

### 1.2.1 A propos des GANs

Introduits par Ian Goodfellow, les GANs produisent des échantillons de données de haute qualité.
<br/>Au cœur de cette architecture se trouve un problème d’optimisation de type min-max impliquant deux réseaux : le générateur
G et le discriminateur T.
<br/>Le générateur et le discriminateur sont entraînés de manière alternée : le générateur cherche à produire des données que le discriminateur ne peut pas distinguer des données réelles, minimisant ainsi la log-probabilité que le discriminateur ait raison.
  
<br/>

### 1.2.2 A propos des Normalizing flows


Les Normalizing Flows (NFs) constituent une catégorie essentielle de modèles génératifs, car ils permettent de suivre la densité des données.
<br/>Cette propriété les rend utiles dans de nombreuses applications, telles que les simulations en physique, la détection d’anomalies, la modélisation du bruit, la génération de sons ou encore les échantillonneurs de Boltzmann.
<br/>En théorie, les Normalizing Flows sont définis comme une bijection, c’est-à-dire une transformation inversible entre l’espace des données X et l’espace latent Z.

## 1.3 Ce qui distingue ces modèles génératifs

Les modèles génératifs ont pour objectif d’apprendre à produire de nouvelles données similaires à celles d’un ensemble d’entraînement (images, texte, sons, etc.).

Ces modèles se différencient principalement par la manière dont ils évaluent la "dissimilarité" entre les données réelles et les données générées.

En d’autres termes, chaque type de modèle génératif (VAE, GAN, diffusion, etc.) choisit sa propre façon de mesurer l’écart entre la distribution des vraies données et celle générée par le modèle.
Cette mesure peut être, par exemple, une divergence de Kullback-Leibler, une distance Wasserstein, ou encore une erreur quadratique moyenne (MSE).

Ainsi, deux modèles génératifs peuvent avoir des architectures, des objectifs et des résultats très différents, simplement parce qu’ils ne mesurent pas la "qualité" d’un échantillon généré de la même manière.
On choisira un modèle plutôt qu'un autre en focntion de notre objectif.

# <font color="#2809a6"> $II$  **Les modèles de diffusion**</font>


## 2.1 Présentation

Les modèles de diffusion tiennent leur nom de leur similitude avec les processus de diffusion en physique qui décrivent comment les molécules se déplacent de zones de grande concentration vers des zones de faible concentration.

En machine learning, les modèles de diffusion utilisent  le concept de dégradation graduelle de la qualité des données dans le but de la reconstruire à l'identique, c’est leur manière d’apprendre à construire une donnée réaliste.
Un modèle de diffusion entraîné peut ensuite construire de nouvelles données à partir de bruit pur donné aléatoirement.

<br/>

### 2.1.1 Différents modèles de diffusion

Il existe divers modèles de diffusion :
* **Modèles probabilistes de débruitage par diffusion (DDPM)** :
le modèle probabiliste de débruitage par diffusion, ou DDPM, fonctionne en éliminant progressivement le bruit d'une image en plusieurs étapes. Il inverse le processus d'ajout de bruit à une image, la rendant de plus en plus nette à chaque étape.

* **Modèles génératifs basés sur des scores** :
les modèles génératifs basés sur des scores apportent une variation aux modèles de diffusion. Ils prédisent la direction à suivre à chaque étape pour aboutir à l'image ou au son final.

* **Modèles de diffusion continus** :
les modèles de diffusion continus se distinguent des autres en ne segmentant pas le processus en étapes discrètes. Ils opèrent en douceur, transformant l'entrée bruyante en une sortie fine-tunée de manière continue.

* **Équations différentielles stochastiques (Score SDEs)** :
les équations différentielles stochastiques de score, ou Score SDEs, sont au cœur de certains modèles de diffusion. Elles apportent une touche d’aléatoire au processus menant au résultat final, en utilisant le calcul stochastique.


Nous nous concentrerons ici sur les modèles probabilistes de débruitage par diffusion , les DDPM.

<br/>

## 2.2 Architecture des modèles de diffusion

Les modèles de diffusion fonctionnent en 2 temps.

Un premier temps d’apprentissage :
* **Processus de diffusion vers l’avant** : introduction de bruit dans des données initiales de manière incrémentale et ceci jusqu’à ce que la donnée initiale soit transformée en bruit pur.

* **Processus de diffusion inverse** : inversion méthodique du processus vers l'avant, sur la donnée devenue bruit pur, pour revenir aux données initiales.

Un second temps de déploiement :
* **Génération** : génération de nouvelles données à partir d'un bruit pur aléatoire. En effet une fois qu’un modèle a appris à revenir aux données initiales à partir de bruit, il saura créer de nouvelles données à partir de bruit pur aléatoire.

Le processus de diffusion vers l’avant ne nécessite aucun apprentissage ni ne représente aucune difficulté, on ne fait qu’ajouter des données à des données initiales. Il est réalisé comme une suite d'étapes de bruitage.

Le processus inverse est lui beaucoup plus complexe, il est réalisé comme une suite d'étapes de débruitage. C'est à ici que le machine learning intervient.

## 2.3 Processus de diffusion vers l’avant (Forward Process)

### 2.3.1 Avant propos

Comme pour tout modèle génératif dont ils font partis, les éléments de base des modèles de diffusion sont :  


* P : une distribution donnée initiale (une image par exemple)
* Q : une distribution latente (une gaussienne standard)
* G : transformation définie par le processus de diffusion

**Nous** **avons** **ainsi** :
P est approximée par une distribution P̂ distribution qui est déterminée par la distribution latente Q et la transformation G.

<br/>

Notons qu’il n’est pas nécessaire de spécifier la fonction G en amont.
Cette fonction sera apprise en minimisant la différence entre P et P̂.

<br/>

### 2.3.2 Principe


La méthode la plus courante du processus de diffusion vers l’avant, consiste à injecter de manière itérative un bruit gaussien jusqu’à ce que l’ensemble de la distribution des données soit gaussien.

D’un point de vue mathématique, ce processus étape par étape est formulé comme une chaîne de Markov : un type de processus stochastique (processus aléatoire qui suit certaines règles probabilistes) pour modéliser des données de séries temporelles séquentielles.

Dans une chaîne de Markov, le résultat à chaque intervalle temporel est influencé uniquement par l’intervalle temporel qui le précède. En d’autres termes, xt, l’état de la chaîne de Markov x à l’intervalle t, est directement influencé uniquement par q($x_{t-1}$). La fonction mathématique définissant la transition depuis n’importe quel q($x_{t}$) vers q($x_{t+1}$) est appelée noyau de transition.

À chaque intervalle temporel t, une petite quantité de bruit gaussien est ajoutée à q($x_{t-1}$). L’image est ensuite redimensionnée pour maintenir une taille d’image constante malgré l’injection continue de pixels aléatoires.

Dans cette formulation, $x_0$ est le point de données propre d’origine ; $x_1$ est le point de données après le premier intervalle temporel, auquel on a ajouté une petite quantité de bruit gaussien ; xT est l’état final du processus de diffusion vers l’avant.

Si T est suffisamment grand, c’est-à-dire après suffisamment d’intervalles, $x_t$ convergera vers un bruit gaussien pur.

On définit la probabilité de transition d’un état $x_{t-1}$ à un état $x_t$ par <font color="#2809a6">q($x_t|x_{t-1}$)</font> pour tout $x \in 1,...,T$.
<br/>
Il s'agit de la prédiction de l’état de la distribution de données q($x_t$), étant donné q($x_{t-1}$).
<br/>
Dans un DDPM standard, ce processus vers l’avant n’implique aucun machine learning. Le résultat final de la chaîne de Markov sera toujours une distribution gaussienne. L’optimisation n’est donc pas nécessaire dans le processus vers l'avant.

À chaque intervalle temporel t, une petite quantité de bruit gaussien est ajoutée à $x_{t-1}$. L’image est ensuite redimensionnée pour maintenir une taille d’image constante malgré l’injection continue de pixels aléatoires.
<br/>Dans cette formulation, $x_0$ est le point de données propre d’origine,  $x_1$ est le point de données après le premier intervalle temporel, auquel on a ajouté une petite quantité de bruit gaussien, $x_{T}$ est l’état final du processus de diffusion vers l’avant.
<br/>
Si T est suffisamment grand, c’est-à-dire après suffisamment d’intervalles, $x_{T}$ convergera vers un bruit gaussien pur.

<br/>

Le bruit gaussien ajouté à chaque intervalle de la chaîne de Markov n’est ni constant ni arbitraire.
<br/>Au lieu de cela, le bruit est dérivé de la structure de l’image d’origine et la vitesse à laquelle il est ajouté augmente régulièrement avec chaque intervalle consécutif.
<br/>La variation de la quantité de bruit améliore à la fois la stabilité de l’entraînement du modèle et la performance générale en équilibrant deux priorités concurrentes.

Un bruit plus important améliore la capacité du modèle à apprendre avec précision dans les zones « à faible densité » des données d’entraînement – les catégories et concepts moins représentés dans les données d’entraînement – en remplissant ces régions avec des données bruyantes.

Cependant, ce procédé comporte un risque d’altération des données, réduisant leur précision globale.

Un bruit plus faible entraîne une corruption de la distribution de données d’origine moindre, mais génère de mauvaises performances sur les régions à faible densité.

Par conséquent, pour obtenir le meilleur des deux mondes, les modèles de diffusion utilisent plusieurs échelles de bruit, c'est-à-dire plusieurs jeux de paramètres, en entraînement.

En effet, des combinaisons différentes des paramètres dont dépendent un bruit gaussien feront varier le bruit gaussien afin de rendre les modèles de diffusion à la fois performants tout en générant des données d’une bonne précision.

### 2.3.3 Bruit et paramétres

Nous avons déjà vu que le processus vers l’avant est un processus faisant transiter une donnée d’un état $x$ à l’instant t vers un état $x+1$ à l’instant t+1 en y ajotant une petite quantité de buit gaussien.

Afin de décrire l'équation mathématiques du passage d'un état $x$ à un état $x+1$, nous allons introduire plusieurs paramètres.

Un bruit gaussien est un type de bruit aléatoire dont les valeurs suivent une distribution normale.
<br/>L’équation mathématique qui décrit sa densité de probabilité est :

<font color="#2809a6"> $f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }$ </font>

Avec :
* x : valeur aléatoire du bruit
* μ : moyenne (espérance mathématique) du bruit
* σ : écart-type du bruit
* $σ^2$: variance du bruit


**La moyenne indique la tendance centrale du bruit.**
<br/>**La variance mesure la propagation ou la dispersion des valeurs du bruit.**

<br/>Un autre paramètre est également important pour décrire un tel buit et notamment gérer ses variations : β.
<br/>**β sert de facteur d’échelle pour la moyenne et la variance du bruit dans la fonction de transition entre 2 étapes**. La modification de la valeur β dans un intervalle donné entraîne une modification du bruit gaussien ajouté à cet intervalle.

β1 est la variance à l’intervalle de temps 1 ; βt est la variance à l’intervalle de temps t, et ainsi de suite, jusqu’à βT. La valeur du taux de β à chaque intervalle est, à son tour, déterminée par la **variance planifiée**.


<br/>À chaque intervalle t, l’image est légèrement décalée par rapport à son itération à l’intervalle précédent (selon la moyenne) et du bruit est ajouté à cette version décalée de l’image (selon la variance). L’ampleur de chaque décalage et de chaque ajout de bruit dépend de la valeur de βt : à mesure que βt augmente, conformément à la variance planifiée, le taux de diffusion augmente de manière constante. β est toujours une valeur comprise entre 0 et 1 : ainsi, 0 < β1 < β2 < ... < βT < 1.

Le choix d’une variance planifiée spécifique pour β est un élément important à prendre en compte. Elle est généralement définie à la main sous la forme d’un hyperparamètre, soit fixe sur une valeur constante, soit en procédant selon une formule avec une valeur de départ et une valeur finale prédéterminées pour β.

La valeur de βt détermine à la fois la moyenne et la variance du bruit gaussien ajouté à l’intervalle t.


* La moyenne μ du bruit gaussien ajouté à l’intervalle temporel t, $μ_t$, est calculée comme suit :
<font color="#2809a6"> $μ_t = (1 - β_t)X_{T-1}$</font>.
 En langage simple, la moyenne du bruit ajouté à chaque intervalle t est en réalité une version à grande échelle de l’image de l’intervalle précédent, $x_{t-1}$. La taille de βt détermine dans quelle mesure cette moyenne s’écarte de l’intervalle précédent : lorsque la valeur βt est très petite, ce décalage est minime, car et le bruit ajouté ressemblera beaucoup à l’image d’origine. À mesure que la valeur de βt augmente, ce décalage devient plus important.


* La variance du bruit gaussien ajouté à l’intervalle de temps t est calculée comme suit : <font color="#2809a6"> $\sum_T = β_tI$</font>, où I est la matrice d’identité. Un βt plus grand produit plus de bruit. Un très petit βt produit un bruit négligeable.

En résumé, à chaque intervalle t, l’image est légèrement décalée par rapport à son itération à l’intervalle précédent (selon la moyenne) et du bruit est ajouté à cette version décalée de l’image (selon la variance). L’ampleur de chaque décalage et de chaque ajout de bruit dépend de la valeur de βt : à mesure que βt augmente conformément à la variance planifiée, le taux de diffusion augmente de manière constante.

Étant donné que l’ajout de bruit gaussien commence progressivement et que le bruit lui-même est toujours dérivé de la structure essentielle de l’image d’origine à l’étape précédente, les qualités essentielles de l’image d’origine sont conservées pendant de nombreuses étapes. Le modèle peut ainsi mieux apprendre les schémas et la structure de la distribution des données d’origine pendant le processus de diffusion inverse.

<br/>

### 2.3.4 Fonction de diffusion vers l'avant


La fonction donnant l'état $x_t$ à partir de l'état $x_{t-1}$ est la suivante :

<font color="#2809a6"> $q_{\phi}(\mathbf{x}_t|\mathbf{x}_{t-1})= \mathcal{N}(\mathbf{x}_t|\sqrt{1- \beta_t}\mathbf{x}_{t-1},\beta_t\mathbf{I})$</font>

<br/>

Avec
* t=1,2,…,T : représente l'étape temporelle
* $β_t$ : constante dépendant de t contrôlant le bruit ajouté à l'étape t et qui vérifie $0 < (1- \beta_t) < 1$
* $\mathbf{I}$ est la matrice identité de covariance
* $\mathcal{N}$(μ,Σ) : une distribution gaussienne de moyenne 𝜇 et de covariance Σ caractérisant le buit ajouté à l'étape t

Le résultat final $x_T$ obtenu après la dernière étape est donc un bruit pur suivant une distribution gaussienne suivant  une loi normale multivariée de moyenne nulle ($\mathbf{0}$) et de matrice de covariance identité ($\mathbf{I}$) : $\mathcal{N}(\mathbf{0},\mathbf{I})$.

<br/>

### 2.3.5 Reparamétrage


L’un des inconvénients de ce processus étape par étape est qu’il est fastidieux et demande énormément de calculs : pour chaque image d’un jeu de données d’entraînement pouvant contenir des milliers ou des millions d’images, le processus vers l’avant nécessiterait des dizaines ou des centaines d’intervalles individuels.

Au lieu d’ajouter du bruit de manière répétée, la formule du processus vers l’avant peut être réécrite en reparamétrant l’expression de <font color="#2809a6">$1 - β_t$</font> comme un nouveau paramètre <font color="#2809a6">$α_t$</font>.

Ce nouveau paramètre peut être étendu à un paramètre supplémentaire,
 <font color="#2809a6">$\tilde{α_t}$</font>, dérivé de la multiplication itérative de <font color="#2809a6">$α_t$</font> à chaque étape de la chaîne jusqu’à ce point.

<br/>

Nous pouvons donc poser :
* $\sigma_t = 1 - \beta_t$
* $\overline{\alpha}_t = \prod_{i=1}^t \alpha_i$

<br/>

### 2.3.6 Fonction de diffusion vers l'avant simplifiée


Étant donné que le processus vers l'avant est une chaîne de Markov avec des transitions gaussiennes, **on peut directement déterminer** $x_t$ **à partir de la donnée initiale** $x_0$ **comme ceci** :

<font color="#2809a6"> $\displaystyle q(\mathbf{x_t}|\mathbf{x_0}) = \mathcal{N}(\mathbf{x_t}|\sqrt{\overline{\alpha}_t}\mathbf{x}_0, (1-\overline{\alpha}_t)\mathbf{I})$</font>

## 2.4 Processus de diffusion inverse (Backward Process)

### 2.4.1 Principe


Dans les modèles de diffusion, le processus de diffusion inverse est l’endroit où le machine learning à proprement dit a lieu. Pour apprendre à effectuer l’inverse des étapes de bruitage du processus vers l’avant, le modèle apprend essentiellement à débruiter le bruit gaussien pur dans une image propre. Une fois le réseau neuronal entraîné, cette capacité peut être utilisée pour générer de nouvelles images à partir d’un bruit gaussien grâce à la diffusion inverse étape par étape.

En théorie, la tâche du modèle peut être considérée comme une simple inversion de la diffusion vers l’avant. Le processus vers l’avant, en commençant par le point de données $x_0$ échantillonné à partir de la distribution de données réelles $q(x)$ du jeu de données entraîné, est défini comme suit : <font color="#2809a6">$ q(x_t | x_{t-1}) $ </font> pour tout $x \in 1,...,T$..

En d'autres termes, étant donné $q(x_{t-1}) $, on peut calculer $q(x_t)$.

La diffusion inverse, serait définie comme suit : $ q(x_{t-1} | x_t) $.

Mais en pratique, ce calcul est irréalisable, **on ne peut pas calculer q de manière exacte, on ne peut que l'approximer**.

<br/>

**Soit donc $p_θ(x)$ un réseau de neuronnes qui approxime $q(x)$, ce réseau de neuronnes est notre modèle de diffusion**.

L’objectif de l’entraînement est donc d’apprendre les paramètres de modèle θ qui font correspondre la sortie de $p_θ(x_{t-1} | x_t )$ à la sortie de $ q(x_{t-1} | x_t)$.

<br/>

Ce modèle, $p_θ(x)$, ne permet pas de prédire directement $x_{t-1}$, ni même le bruit spécifique ajouté entre $x_{t-1}$ et $x_t$.

Au lieu de cela, **il prédit l’ensemble du bruit présent dans $x_t$, puis élimine une fraction de ce bruit (en fonction de l’état de la variance planifiée à cette étape) pour obtenir $x_{t-1}$**.

On peut simplifier encore ce processus en estimant uniquement la moyenne du bruit gaussien, ou bien encore en prédisant la variance.

Rappelons à nouveau que la moyenne du bruit gaussien ajouté dans la diffusion vers l’avant n’est pas arbitraire : bien qu’elle soit en effet aléatoire, la structure du bruit est initialement dérivée de la structure de l’image d’origine $x_0$. Par conséquent, en apprenant à prédire avec précision le bruit grâce à la diffusion inverse, le modèle apprend non seulement à débruiter l’image, mais il apprend également implicitement la structure de $x_0$.

<br/>

### 2.4.2 Calcul du modèle de diffusion


Nous venons de voir que notre fonction de processus inverse était en réalité un modèle de diffusion déterminé par une phase d'entrainement classique des réseaux de neuronnes.

<font color="#2809a6">$\displaystyle q(\mathbf{x}_{t-1}|\mathbf{x}_t,\mathbf{x}_0)=\mathcal{N}(\mathbf{x}_{t-1}|\tilde{\mu}_t(\mathbf{x}_t,\mathbf{x}_0),\tilde{\beta}_t\mathbf{I})$</font>

Avec
* $\displaystyle \tilde{\mu}_t(\mathbf{x}_t,\mathbf{x}_0)=\frac{(1-\alpha_t)\sqrt{\overline{\alpha}_{t-1}}}{1-\overline{\alpha}_t} \mathbf{x}_0 + \frac{(1-\overline{\alpha}_{t-1})\sqrt{\alpha_t}}{1-\overline{\alpha}_t} \mathbf{x}_t$
* $\displaystyle \tilde{\beta}_t=\frac{(1-\alpha_t)(1-\overline{\alpha}_{t-1})}{1-\overline{\alpha}_t}$

<br/>

Nous pouvons donc définir $p_{\theta}(\mathbf{x}_{t-1}|\mathbf{x}_t)$ comme ceci :

<font color="#2809a6">$\displaystyle p_{\theta}(\mathbf{x}_{t-1}|\mathbf{x}_t) \stackrel{\mathrm{def}}{=} \mathcal{N}(\mathbf{x}_{t-1}|\mu_{\theta}(\mathbf{x}_t, t),\sigma_t^2 \mathbf{I}))$ </font>

Avec :
* $\mu_{\theta}(\mathbf{x}_t, t)$ is a neural network (or a function) parameterized by $\theta$
* $\displaystyle \sigma_t^2 = \frac{(1-\alpha_t)(1-\overline{\alpha}_{t-1})}{1-\overline{\alpha}_t}$.

### 2.4.3 Optimisation du modèle de diffusion


Dans le domaine du machine learning, on utilise une fonction de perte pour optimiser la performance d'un modèle en mesurant l'écart entre ses prédictions et celles de la « vérité terrain ». Optimiser un modèle consiste à ajuster ses paramètres pour minimiser le résultat de cette fonction de perte.

Une fonction de perte est un type de fonction objective dont la minimisation est l'objectif de l'entraînement de notre modèle.
Elle quantifie la différence (« perte ») entre une valeur prédite (la sortie du modèle) pour une entrée donnée et la valeur réelle, ou « vérité terrain ».

Plus la perte est faible et plus les prédictions sont correctes. En cherchant le minimum de cette fonction, on obtient la meilleure approximation possible de nos données réelles par nos prédictions.
Le choix de la fonction de perte est donc fondamentale pour ajuster au mieux les paramètres de notre modèle et approximer le plus possible les données réelles.

<br/>

Dans le cas des modèles de diffusion, on ne cherche pas directement à minimiser une fonction de perte.
<br/>À la place, on essaie d’optimiser une approximation appelée "limite inférieure variationnelle" (VLB), qui regroupe plusieurs termes de perte.

Ces termes de perte mesurent la différence entre deux distributions de probabilités, grâce à un outil appelé divergence de Kullback-Leibler (KL). Cela permet par exemple de comparer comment les pixels sont distribués dans deux images.

La fonction de perte totale est composée de trois parties :

* **LT** : compare la distribution du bruit final (après avoir ajouté du bruit plusieurs fois) à ce que le modèle pense qu’il devrait obtenir.  Ce terme est souvent négligé, car ce bruit final suit une loi connue (gaussienne) et ne dépend pas vraiment des paramètres du modèle.

* **Lt** : compare, à chaque étape, la vraie façon dont le bruit a été ajouté à ce que le modèle prédit pour l’enlever. C’est le cœur de l’apprentissage du modèle : apprendre à bien débruiter étape par étape.

* **L0** : mesure à quel point l’image finale générée par le modèle est proche de l’image originale.C’est une mesure de "vraisemblance" : le modèle essaie de prédire une image finale qui ressemble le plus possible à une vraie.

<br/>

En pratique, cette fonction de perte peut être simplifiée à une erreur quadratique moyenne (MSE) entre :

* le bruit que le modèle prédit, et
* le vrai bruit ajouté à l’image au départ.

<br/>

La VLB peut être simplifiée sous la forme d’une erreur quadratique moyenne (MSE) entre le bruit que le modèle prédit ($\varepsilon_θ$) et le bruit réellement ajouté pendant le processus de bruitage ($\varepsilon$), à chaque étape.
C’est pour cela que le modèle ne génère pas directement l’image finale, mais apprend plutôt à prédire le bruit ajouté à chaque étape.

Pendant l’entraînement, on calcule le gradient de cette fonction de perte par rétropropagation, puis on ajuste les poids du modèle avec une méthode comme la descente de gradient. Cela permet au modèle d’améliorer peu à peu ses prédictions sur l’ensemble des images d'entraînement.

## 2.6 Génération d’images avec des modèles de diffusion


Une fois que le modèle de diffusion a appris à estimer avec précision le bruit à soustraire à chaque intervalle, il peut servir à générer de nouvelles images en échantillonnant une image aléatoire avec du bruit $x_T$ à partir de la distribution de données qu’il a apprise et en la débruitant pendant les intervalles T.

L'introduction d’un élément légèrement aléatoire dans le processus d’échantillonnage permet aux modèles de diffusion de produire de nouvelles images qui ressemblent aux données d’entraînement, plutôt que de reproduire directement les images d’entraînement.

Contrairement au processus d'entraînement par diffusion inverse, le nombre d’intervalles au sein du processus de génération ne doit pas nécessairement correspondre au nombre d’intervalles utilisés dans le processus vers l’avant. Cela est rendu possible grâce au modèle, entraîné à prédire l’ensemble du bruit à chaque étape de l’image, plutôt que la quantité spécifique de bruit à supprimer sur cet intervalle.

Le nombre réduit d’intervalles implique une plus grande vitesse et des demandes de calcul réduites, avec un compromis potentiel sur les plus petits détails ; les intervalles supplémentaires améliorent généralement la précision, mais au prix d’une réduction de la vitesse et d’une augmentation des coûts de calcul.



# <font color="#2809a6"> $III$ **Evaluation des modèles de diffusion**</font>


Pour une évaluation fiable des modèles génératifs, il est essentiel que les indicateurs utilisés soient indépendants du modèle (model-agnostic).
La méthode et l’algorithme de calcul de ces indicateurs d’évaluation doivent être identiques pour tous les modèles génératifs.

De plus, pour être facilement calculables, ces indicateurs doivent dépendre uniquement d’un ensemble d’échantillons tirés à la fois de $P$ (la distribution réelle) et de $\hat{P}$
(la distribution générée), sans nécessiter d’entraînement supplémentaire.

À mesure que les performances des modèles génératifs s'améliorent, le besoin d'indicateurs d’évaluation plus fins devient tout aussi essentiel que prioritaire.

## 3.1 Qualité contre diversité


La comparaison des différents modèles génératifs a mis en avant le duel entre qualité et diversité dans les données produites par tous les modèles.

Les modèles performants d'un point de vue qualité, c'est-à-dire produisant des données très vraisemblables et réalistes, sont ceux qui n'offrent qu'une variété d'image très limitée.

Et inversement les modèles apportant une grande richesse de diversité des données produites, ne sauront pas produire des données de grande qualité, les incohérences dans les données produites seront facilement visibles à l'œil nu.

Bien que les nouveaux modèles tendent à diminuer cette dualité entre qualité et diversité, les modèles performants et de haute qualité restent malgré tout d'une expressivité quelque peu limitée et contrainte.
En fonction du domaine d'application des modèles et de leur objectifs, on va davantage se tourner vers des modèles offrant plus de qualité au prix d'une faible diversité, dans le domaines de la génération d'images de haute résolution,  ou inversement dans les domaine de la génération de molécules médicamenteuses ou la détection d'anomalie.




<br/>

## 3.2 Métriques traditionnels


Les métriques Inception Score (IS) et Fréchet Inception Distance (FID) ont été les premiers métriques évaluant les modèles génératifs.

<br/>

L'**Inception Score** est une métrique conçue pour évaluer à la fois la qualité et la diversité des images générées par un modèle génératif.
L’IS repose sur la capacité de classification du modèle Inception.
Les échantillons générés doivent pouvoir être facilement classifiés par le modèle Inception, avec une distribution des étiquettes similaire à celle des données réelles.

L’Inception Score présente plusieurs limites importantes dont les principales sont les suivantes :

* **Insensibilité à la diversité intra-classe** : si un modèle génère une seule image de haute qualité par classe, le score IS reste élevé, même si la diversité est faible.

* **Mesure indirecte du réalisme** : des images floues, bruitées ou déformées peuvent obtenir un bon score si la confiance en leur classification est élevée.

* **Comparaison inadéquate des distributions** : l’IS ne compare pas directement la distribution des images générées à celle des données réelles. Ainsi, même lorsque les distributions générée  $\hat{P}$ et réelle $P$ sont identiques, le score IS n’est pas nécessairement optimal.

<BR/>

Pour toutes ces raisons, l’Inception Score a été progressivement remplacée à partir de 2020 par la Fréchet Inception Distance (FID), qui est jugée mieux corrélée à la perception humaine de la qualité.

<BR/>

La **Fréchet Inception Distance** mesure la distance entre les vecteurs de caractéristiques latentes extraits des images réelles et générées à l’aide de leur moyenne et covariance.
La FID est basée sur la distance de Fréche.

Plus la FID est faible, plus les images générées ressemblent statistiquement aux vraies images en termes de qualité visuelle et diversité.

La présente elle aussi des limites en terme d'évaluation de la qualité et de la dievrsité.

<BR/>

* **Comparaison uniquement des résumés statistiques** la moyenne et la covariance, par conséquent, il peut ignorer certains aspects visuels importants, comme la texture ou la structure locale, pourtant bien perçus par les humains.

* **Non distinction des types d’erreurs visuelles**, par exemple, une image légèrement bruitée et une image représentant un objet complètement incorrect peuvent obtenir un score similaire, ce qui ne reflète pas le jugement humain.

<br/>

Que ce soit l'Inception Score ou la Fréchet Inception Distance, la qualité et la diversité ne sont pas simultanément pas évaluées suffisamment précisément. Pour l'évaluation de modèles génératifs avancés, ces métriques n'offrent pas toujours des performances suffisament nuancées et complètes.

<br/>

Il faut donc utliser d'autres métriques afin de pouvoir comparer et évaluer les modèles génératifs les plus récents et les plus performants.
Inspirées de la classification, 2 notions ont émergées afin de déterminer la qualité et la diversité d'un modèle génératif : la précision (la capacité à ne pas étiqueter à tort les éléments comme positifs), et le rappel (la capacité à détecter tous les éléments positifs réels).


<br/>

## 3.3 Notion de précision et rappel

### 3.3.1 Précision et rappel pour les classifications classiques


Avant de présenter leur adaptation aux modèles génératifs et d'expliquer en quoi ces métriques sont performants dans l'évaluation des modèles génératifs les plus avancés, rappelons leurs définitions classiques dans les tâches de classification.

<br/>

La **précision** est la proportion de points de données prédits comme positifs qui sont correctement classés.

Le **rappel** est la proportion de points de données réellement positifs qui sont correctement classés. Formellement :

<br/>

Dans une tâche de classification binaire, posons les éléments suivants :  
* P le nombre d’instances positives,
* PP le nombre d’instances prédites comme positives, et
* TP le nombre de vrais positifs.

La précision et le rappel en classification binaire sont alors définis comme suit :

* $Précision = \frac{TP}{PP}$

* $Rappel = \frac{TP}{P}$

<br/>

**La précision et le rappel permettent ainsi d’évaluer indépendamment la qualité et la diversité.**

<br/>

### 3.3.3 Précision et rappel pour les modèles génératifs


Dans les modèles génératifs, l’étiquette positive fait référence au fait qu’un point peut être généré par $P$, et les points de données prédits comme positifs sont ceux qui peuvent être générés par $\hat{P}$.

Avec cette transposition, on peut comprendre ce que signifient la précision et le rappel dans le cadre de la modélisation générative :

* La précision mesure la proportion d’échantillons générés qui pourraient l’être par $P$. Une haute précision dans un modèle génératif indique que les échantillons générés sont de haute qualité.
* Le rappel évalue la proportion d’échantillons tirés de $P$ qui peuvent être générés par $\hat{P}$. Un rappel élevé suggère que les points de données générés par $\hat{P}$ sont très diversifiés.

Cependant, cette transposition à la modélisation générative n’est pas immédiate. La difficulté d’appliquer la précision et le rappel aux modèles génératifs réside dans la définition du fait qu’un point peut être généré par les distributions $P$ ou $\hat{P}$.

Contrairement aux tâches de classification, où les étiquettes de vérité terrain sont connues, dans les tâches génératives, il n’existe pas d’étiquette explicite indiquant si un point est issu d’une distribution donnée.

Ainsi, les différentes définitions varient selon la manière dont cette notion est transposée à la tâche générative.


<br/>

## 3.4 Métrique alternative : Densité et couverture


La densité et la couverture constituent une métrique alternative à la précision et au rappel.

En effet, la précision et le rappel étant sensibles aux valeurs aberrantes (outliers), les auteurs cherchent à proposer une évaluation plus robuste de la fidélité et de la diversité. Ces métriques n’estiment pas dans quelle mesure les supports des distributions se recouvrent, mais plutôt dans quelle mesure les échantillons générés occupent les voisinages des points de données réels. Pour cette raison, elles ne relèvent pas de la catégorie des métriques de précision/rappel et sont appelées métriques de densité et de couverture.

Ces métriques reposent sur l’estimation des voisinages des points de données réels.

La densité et la couverture sont conçues pour évaluer indépendamment la qualité et la diversité des échantillons générés.

Densité : Cette métrique évalue la qualité. Elle quantifie à quel point les échantillons générés remplissent de manière dense les régions autour des points de données réels. Contrairement à la précision, qui se contente de vérifier si un échantillon généré appartient au support des données réelles, la densité mesure la concentration des échantillons générés dans le voisinage des points de données réels.

Couverture : Cette métrique évalue la diversité. Elle quantifie la proportion des voisinages des points de données réels dans lesquels il existe au moins un point généré.


Inconvénients :
Bien que la densité et la couverture offrent une évaluation plus robuste, leur complexité peut les rendre moins interprétables comparativement à la simplicité des métriques de précision et rappel.
De manière similaire à l’approche basée sur le support utilisée pour la précision et le rappel, l’efficacité de ces métriques dépend fortement du choix de l’hyperparamètre K dans l’algorithme des k plus proches voisins (k-NN).

<br/>

## 3.5 Métrique alternative : Couverture Précision-Rappel


Couverture Précision-Rappel, une métrique récente introduite par Cheema et Urner.
Cette métrique vise à combler le fossé entre les concepts théoriques et le calcul pratique de la précision et du rappel. L’objectif est de formuler une définition de la précision et du rappel plus proche du calcul pratique.

Il s'agit d'un compromis entre précision et rappel pour les modèles de diffusion.

## 4.1 Avantages des Modèles de Diffusion

*  **Stabilité de l'entraînement** : Comparés aux GANs, les modèles de diffusion tendent à être plus stables à l’entraînement. Ils ne sont pas sujets à des problèmes comme le mode collapse que l’on rencontre souvent avec les GANs.
*  **Capacité à générer des images de haute qualité** : Les modèles de diffusion produisent souvent des images plus nettes et plus réalistes grâce à leur capacité à modéliser progressivement la distribution des données.
*  **Flexibilité** : Ces modèles sont adaptables et peuvent être utilisés pour une grande variété de types de données.

<br/>

## 4.2 Inconvénients des Modèles de Diffusion

*  **Coût computationnel** : L'un des principaux inconvénients est le coût en calcul. En raison de la nécessité de simuler plusieurs étapes du processus de diffusion inverse, les modèles de diffusion nécessitent souvent des ressources informatiques considérables pour l'entraînement et la génération.
*  **Latence de génération** : Comparés à d'autres modèles génératifs comme les GANs, les modèles de diffusion peuvent être plus lents à générer des échantillons, car le processus inverse peut prendre plusieurs étapes.

# <font color="#2809a6"> $V$ **Challenges et défis**</font>


Afin que les modèles de diffusion ouissent se développer et tenir leurs promesses, il faut encore répondre aux nombreux défis qu'ils posent encore actuellement

* **Temps de génération long** :
Génération lente à cause du grand nombre d'étapes nécessaires (centaines à milliers).

* **Coût de calcul élevé** :
Le débruitage itératif consomme beaucoup de ressources GPU/TPU à l'entraînement et à l'inférence.

* **Compromis qualité/diversité** :
Difficile d'équilibrer réalisme et variété ; les sorties peuvent être trop moyennées.

* **Contrôle limité de la génération** :
Guider précisément le processus (texte, esquisse, etc.) est complexe et nécessite des architectures spécialisées.

* **Apprentissage instable** :
L'entraînement est sensible, demande beaucoup de données et un bon prétraitement.

* **Évaluation difficile** :
Les métriques classiques (FID, IS) ne suffisent pas pour bien juger qualité et diversité.


<br/>

Afin de répondre à ces défis, de nombreuses recherches sont en cours afin d'améliorer les modèles de diffusion selon les axes suivants :

* **Réduction du nombre d'étapes** :
Nouvelles méthodes (DDIM, DPM-Solver, FastDPM) permettent une génération en 4 à 50 étapes (vs ~1000), pour accélérer l'inférence sans sacrifier la qualité.

* **Distillation et compression** :
Techniques comme Progressive Distillation ou Latent Consistency Models transforment les modèles lents en versions rapides (1 à quelques étapes).

* **Guidage conditionnel renforcé** :
Amélioration du contrôle via texte, masque, image, etc.
Exemples : ControlNet, Prompt-to-Prompt, Classifier-free guidance.

* **Utilisation d’espaces latents** :
Travailler en espace latent (ex. Stable Diffusion) permet de réduire la complexité et d’accélérer la génération.

* **Architectures plus efficaces** :
Adoption de modèles avancés : U-ViT, UNet optimisés, Transformers pour de meilleures performances.

* **Évaluation plus pertinente** :
Développement de métriques mieux alignées avec la perception humaine pour évaluer qualité et diversité.

* **Applications multi-modales** :
Diffusion élargie à d'autres domaines : audio (MusicLM), vidéo, 3D, code, NLP, etc.
